```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
FEATURES_PATH = 'ames_features.csv'

df = pd.read_csv(FEATURES_PATH)
print(f'Loaded: {df.shape[0]:,} rows x {df.shape[1]} columns')
```

## **Math Task 1 — Manual Mean & Standard Deviation**

---
# **Compute the mean and standard deviation of the target manually using NumPy (not pandas.mean())**


---


Computing mean and std of SalePrice from scratch using NumPy
to verify we understand the formulas before using built-in functions.

```python

# Source: Week 8, Day 36 — Vectors & NumPy Operations

prices = np.array(df['SalePrice'])
# Manual mean
mean_manual = np.sum(prices) / len(prices)
# Manual std:
std_manual = np.sqrt(np.sum((prices - mean_manual) ** 2) / len(prices))
# Compare with NumPy built-in
mean_numpy = np.mean(prices)
std_numpy  = np.std(prices)
print(f'Manual mean : ${mean_manual:,.2f}')
print(f'NumPy  mean : ${mean_numpy:,.2f}')
print(f'Manual std  : ${std_manual:,.2f}')
print(f'NumPy  std  : ${std_numpy:,.2f}')
```

## **Math Task 2 — Manual Standardisation vs StandardScaler**

---
# **Standardise one column by hand using broadcasting: z = (X - mean) / std — compare with StandardScaler output**


---


Standardising Overall Qual by hand using broadcasting: z = (X - mean) / std
then comparing result with sklearn StandardScaler to confirm they match.

```python
# Source: Week 8, Day 39 — Broadcasting & Standardization
col = np.array(df['Overall Qual'])
mean_col = np.mean(col)
std_col  = np.std(col)
z_manual = (col - mean_col) / std_col
scaler   = StandardScaler()
z_scaler = scaler.fit_transform(col.reshape(-1, 1)).flatten()
print(f'Mean of col     : {mean_col:.4f}')
print(f'Std of col      : {std_col:.4f}')
print(f'\nFirst 5 — Manual : {z_manual[:5].round(4)}')
print(f'First 5 — Scaler : {z_scaler[:5].round(4)}')
print(f'\nMax difference   : {np.abs(z_manual - z_scaler).max():.10f}')
```

## **Math Task 3 — Cosine Similarity**

---
# **Compute cosine similarity between the feature vectors of the highest-value and lowest-value records**


---


Measuring the angle between the feature vectors of the most expensive
and least expensive house. A value near 1.0 means they point in the
same direction despite very different magnitudes.

```python
# Source: Week 8, Day 37 — Dot Product & Cosine Similarity
numeric_df = df.select_dtypes(include=[np.number])
# Get the row index of highest and lowest SalePrice
idx_high = df['SalePrice'].idxmax()
idx_low  = df['SalePrice'].idxmin()
vec_high = numeric_df.loc[idx_high].values.astype(float)
vec_low  = numeric_df.loc[idx_low].values.astype(float)
# Cosine similarity = dot product / (norm_a * norm_b)
dot_product   = np.dot(vec_high, vec_low)
norm_high     = np.linalg.norm(vec_high)
norm_low      = np.linalg.norm(vec_low)
cosine_sim    = dot_product / (norm_high * norm_low)

print(f'Highest SalePrice : ${df.loc[idx_high, "SalePrice"]:,.0f}')
print(f'Lowest  SalePrice : ${df.loc[idx_low,  "SalePrice"]:,.0f}')
print(f'\nCosine Similarity : {cosine_sim:.4f}')
print(f'Interpretation    : {"similar" if cosine_sim > 0.9 else "different"} feature profiles')
```

## **Math Task 4 — Probability Estimation**

---
# **Estimate a probability: e.g., what fraction of high-quality items have a target above a certain threshold**


---


Estimating: what fraction of high-quality houses (Overall Qual >= 7)
have a SalePrice above the dataset median?
This links quality to price in a probabilistic way.

```python
# Source: Week 8, Day 37 — Probability Estimation
median_price  = np.median(df['SalePrice'])
high_qual     = df[df['Overall Qual'] >= 7]

above_median  = (high_qual['SalePrice'] > median_price).sum()
total_high    = len(high_qual)
probability   = above_median / total_high

print(f'Median SalePrice           : ${median_price:,.0f}')
print(f'High quality houses (>= 7) : {total_high}')
print(f'Above median price         : {above_median}')
print(f'\nP(price > median | quality >= 7) = {probability:.4f} ({probability*100:.1f}%)')
```
